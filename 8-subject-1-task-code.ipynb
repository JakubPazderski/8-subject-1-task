```python
import pandas as pd

data=pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/',header=0)

df=data[0]

df.rename(columns={'TITLE':'TYTUŁ','ARTIST':'ARTYSTA','YEAR':"ROK","HIGH POSN":'MAX POZ'},inplace=True)

unique_artists=df['ARTYSTA'].unique()

count_of_unique_artists=len(unique_artists)
#Na liście znajduje się 47 pojedynczych artystów

the_most_artist=df['ARTYSTA'].value_counts()
#Na liście najczęściej pojawia się COLDPLAY oraz TAKE THAT

df['TYTUŁ']=df['TYTUŁ'].str.capitalize()
df['ARTYSTA']=df['ARTYSTA'].str.capitalize()

df.drop('MAX POZ',axis=1,inplace=True)

df['ROK'].value_counts()
#Najwięcej albumów wyszło w 2000 roku

mask=(df['ROK']>=1960)&(df['ROK']<=1990)
len(df[mask])
#Na liście znajdują się 22 albumy wydane między 1960 a 1990 rokiem.

sorteddata=df.sort_values(by='ROK')
#Najmłodszy album na liście został wydany w 1965 roku

final_list=sorteddata.drop_duplicates(subset=['ARTYSTA'])


    


```
